```python
import torch
import torch.nn as nn
```

```python
class NeuralNet(nn.Module):
    def __init__(self):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(1, 8)
        self.fc2 = nn.Linear(8, 4)
        self.output = nn.Linear(4, 1)
    
    def forward(self, x):
        x = self.fc1(x)
        x = self.fc2(x)
        x = self.output(x)
        return x
```

```python
ann = NeuralNet()
```

```python
criterion = nn.MSELoss()
optim = torch.optim.Adam(ann.parameters(), lr=0.01)
```

```python
X_data = torch.Tensor([[-10], [-5], [9], [100]])
y_data = torch.Tensor([[-88], [-43], [83], [902]])
```

```python
for epoch in range(500):
    optim.zero_grad()
    y_hat = ann(X_data)
    loss = criterion(y_hat, y_data)
    loss.backward()
    optim.step()
    
    print('Epoch : ', epoch, ' -'*10, ' Loss : ', loss.item())
```

```python
x = torch.Tensor([-80])
print(ann(x))
```

```python
print('Weight FC 1 \n', ann.fc1.weight, '\n', ann.fc1.weight.size())
print('Weight FC 2 \n', ann.fc2.weight, '\n', ann.fc2.weight.size())
print('Weight Output \n', ann.output.weight, '\n', ann.output.weight.size())
```
