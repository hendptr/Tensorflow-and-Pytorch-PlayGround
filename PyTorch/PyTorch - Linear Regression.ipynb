```python
import torch
import torch.nn as nn
import torch.optim
```

```python
class SingleLinear(nn.Module):
    def __init__(self):
        super(SingleLinear, self).__init__()
        self.W = torch.randn(1, requires_grad=True)
        self.B = torch.randn(1, requires_grad=True)
    def forward(self, x):
        return x * self.W + self.B
```

```python
linear = SingleLinear()
```

```python
X_data = torch.arange(-100, 100, dtype=torch.float)
y_data = X_data * 9 + 2
```

```python
optim = torch.optim.Adam([linear.W, linear.B], lr=0.1)
loss = nn.MSELoss()
```

```python
for epoch in range(1000):
    optim.zero_grad()
    y_pred = linear(X_data)
    cur_loss = loss(y_pred, y_data)
    cur_loss.backward()
    optim.step()
    print('Epoch : ', epoch, ' -'*10, ' Loss :', cur_loss.item(), ' W : ', linear.W.item(), ' B : ', linear.B.item())
```

```python
x = torch.Tensor([-100])
print(linear(x))
```
