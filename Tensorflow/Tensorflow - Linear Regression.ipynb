```python
import tensorflow as tf
from tensorflow.keras import Model, Sequential
from tensorflow.keras.layers import Dense
```

```python
class LinearReg(Model):
    def __init__(self):
        super(LinearReg, self).__init__()
        self.W = tf.Variable(1.0)
        self.b = tf.Variable(1.0)
    def __call__(self, x):
        y = x * self.W + self.b
        return y
```

```python
Reg = LinearReg()
```

```python
x_data = tf.random.normal(shape=[1000])
noise = tf.random.normal(shape=[1000])
y_data = x_data * -4 + 100 + noise
import matplotlib.pyplot as plt
plt.scatter(x_data, y_data)
```

```python
criterion = tf.keras.losses.MeanSquaredError()
optim = tf.keras.optimizers.Adam(lr=0.001)
```

```python
for epoch in range(1500):
    with tf.GradientTape() as tape:
        y_pred = Reg(x_data)
        loss = criterion(y_pred, y_data)
    gradient = tape.gradient(loss, Reg.trainable_variables)
    optim.apply_gradients(zip(gradient, Reg.trainable_variables))
    print('Epoch {} ---------- Loss {} ------ W {:0.4f} --------- B {:0.4f}'.format(epoch+1, loss.numpy(), Reg.W.numpy(), Reg.b.numpy()))
```

```python
plt.figure()
plt.scatter(x_data, y_data, c='b', label='Original data')
plt.plot(x_data, Reg(x_data), c='r', label='Prediction')
plt.legend()
plt.show()
```

```python
Reg.trainable_variables
```

```python
model = Sequential()
model.add(Dense(64))
model.add(Dense(32))
model.add(Dense(1))
model.compile(loss='MSE', optimizer='Adam')
```

```python
model.fit(x_data, y_data, epochs=1500)
```

```python
plt.figure()
plt.scatter(x_data, y_data, c='b', label='Original data')
plt.plot(x_data, model.predict(x_data), c='r', label='Prediction NN')
plt.plot(x_data, Reg(x_data), c='y', label='Prediction ')
plt.legend()
plt.show()
```
